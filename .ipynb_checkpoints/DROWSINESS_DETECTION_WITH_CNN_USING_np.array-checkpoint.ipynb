```python
class convolution_layer_cnn: #wriitng the fucntion for conv layer. that perfomrs the convolve opertaion
    def __init__(self, filt_num, dim_fil):
        self.filt_num = filt_num
        self.dim_fil = dim_fil
        self.fil_convolut = np.random.randn(filt_num, dim_fil, dim_fil)/(dim_fil * dim_fil)

    def size_img(self, img_var):
        h, w = img_var.shape
        self.img_var = img_var
        for j in range(h - self.dim_fil + 1):
            for k in range(w - self.dim_fil + 1):
                img_var_patch = img_var[ j : (j + self.dim_fil), k : (k + self.dim_fil)]
                yield img_var_patch, j, k

    def prp_fow(self, img_var):
        h, w = img_var.shape
        conv_out = np.zeros((h - self.dim_fil + 1, w - self.dim_fil + 1, self.filt_num))
        for img_var_patch, i, j in self.size_img(img_var):
            conv_out[i, j] = np.sum(img_var_patch * self.fil_convolut, axis = (1,2))
        return conv_out

    def prp_bck(self, out_lay, alpha_lr):
        par = np.zeros(self.fil_convolut.shape)
        for img_var_patch, i, j in self.size_img(self.img_var):
            for k in range(self.filt_num):
                par[k] += img_var_patch*out_lay[i, j, k]

            self.fil_convolut -= alpha_lr*par
            return par
```

```python
class max_pooling_layer: #function that defines a max pol operation 
    def __init__(self, dim_fil):
        self.dim_fil = dim_fil

    def size_img(self, img_var):
        new_height = img_var.shape[0] // self.dim_fil
        new_width = img_var.shape[1] // self.dim_fil
        self.img_var = img_var

        for i in range(new_height):
            for j in range(new_width):
                img_var_patch = img_var[(i*self.dim_fil) :(i*self.dim_fil + self.dim_fil), (j*self.dim_fil) : (j*self.dim_fil + self.dim_fil)]
                yield img_var_patch, i, j

    def prp_fow(self, img_var):
        h, w, filt_num = img_var.shape
        res = np.zeros((h // self.dim_fil, w // self.dim_fil, filt_num))

        for img_var_patch, i, j in self.size_img(img_var):
            res[i, j] = np.amax(img_var_patch, axis = (0,1))

        return res

    def prp_bck(self, out_lay):
        dl_dmax_pooling_layer = np.zeros(self.img_var.shape)
        for img_var_patch, i, j in self.size_img(self.img_var):
            h, w, filt_num = img_var_patch.shape
            maxv = np.amax(img_var_patch, axis=(0,1))

            for i1 in range(h):
                for j1 in range(w):
                    for k1 in range(filt_num):
                        if img_var_patch[i1, j1, k1] == maxv[k1]:
                            dl_dmax_pooling_layer[i*self.dim_fil + i1, j*self.dim_fil + j1, k1] = out_lay[i, j, k1]

            return dl_dmax_pooling_layer
```

```python
class softmax_layer_final: #function that defines the activation function 
    def __init__(self, input_node, softmax_layer_final_node):
        self.weight = np.random.randn(input_node, softmax_layer_final_node)/input_node
        self.bias = np.zeros(softmax_layer_final_node)

    def prp_fow(self, img_var):
        self.original_img_var_shap = img_var.shape
        img_var_modified = img_var.flatten()
        self.inp_changed = img_var_modified
        res_val = np.dot(img_var_modified, self.weight) + self.bias
        self.out_lay = res_val
        exp_out_lay = np.exp(res_val)

        return exp_out_lay/np.sum(exp_out_lay, axis = 0)

    def prp_bck(self, out_lay, alpha_lr):
        for i, grad in enumerate(out_lay):
            if grad == 0:
                continue

            transformation_eq = np.exp(self.out_lay)
            s_total = np.sum(transformation_eq)

            var1 = -transformation_eq[i]*transformation_eq / (s_total ** 2)
            var1[i] = transformation_eq[i] * (s_total - transformation_eq[i]) / (s_total ** 2)

            var_2 = self.inp_changed
            var_3 = 1
            var_3_inp = self.weight

            var_4 = grad * var1

            var_5 = var_2[np.newaxis].T @ var_4[np.newaxis]
            var_6 = var_4 * var_3
            var_6_inp = var_3_inp @ var_4

        self.weight -= alpha_lr * var_5
        self.bias -= alpha_lr * var_6

        return var_6_inp.reshape(self.original_img_var_shap)
```

### Building the model 

```python
convolution_layer = convolution_layer_cnn(10, 7)
max_pooling_var = max_pooling_layer(4)
soft_max = softmax_layer_final(10*10*10, 2)  #our model has used 1 input, 1 hidden layer that has (1 conv, 1 max pooling, 1 activation function) 
```

```python
def cnn_forward_prop(img_var, label): #will take input images , class label and will perform forward propogation operation by calling the above mentioned functions. 
    out_lay_p = convolution_layer.prp_fow((img_var / 255) - 0.5)
    out_lay_p = max_pooling_var.prp_fow(out_lay_p)
    out_lay_p = soft_max.prp_fow(out_lay_p)

    cross_ent_loss = -np.log(out_lay_p[label])
    accuracy_eval = 1 if np.argmax(out_lay_p) == label else 0

    return out_lay_p, cross_ent_loss, accuracy_eval

def training_cnn(img_var, label, alpha=0.001): #we calculate the loss with a ;earning rate of 0.001
    out_lay, loss, acc = cnn_forward_prop(img_var, label)

    gradient = np.zeros(7)
    gradient[label] = -1 / out_lay[label]

    backward_com_grad = soft_max.prp_bck(gradient, alpha)
    backward_com_grad = max_pooling_var.prp_bck(backward_com_grad)
    backward_com_grad = convolution_layer.prp_bck(backward_com_grad, alpha)

    return loss, acc
```

```python
print("-------------Train----------------")
for epoch1 in range(10):
    print('The epoch number is %d ----> ' %(epoch1 + 1))

    seed = np.random.permutation(len(tr_img))
    tr_img = tr_img[seed]
    tr_lbl = tr_lbl[seed]

    loss = 0
    val_num = 0

    for i, (grayscale_img_var, label) in enumerate(zip(tr_img, tr_lbl)):
        if i % 100 == 0:
            print(" For step count 100 %d is : avg Loss %.3f and acc: %d%%" %(i+1, loss / 100, val_num))
            loss = 0
            val_num = 0

        loss_val_1, accura = training_cnn(grayscale_img_var, label)
        loss += loss_val_1
        val_num += accura

print("-----------------Test-------------------")
loss = 0
val_num = 0

for grayscale_img_var, label in zip(tt_img, tt_lbl):
    _, loss_val_1, accura = cnn_forward_prop(grayscale_img_var, label)
    loss += loss_val_1
    val_num += accura

num_tests = len(tt_img)
print('Test Loss:', loss / num_tests)
print('Test Accuracy:', val_num / num_tests)
```

```python
def predict(x):
    new_img_var = []
    loss_array = []
    accuracy_array = []
    dictionary_cont_dat = {}
    class_val = {0: 'Closed', 1: 'Open'}
    for data in class_val.keys():
        new_img_var.append(x[0]) 
    #print(new_image)
    for grayscale_img_var, label in zip(new_img_var, class_val.keys()):
        loss = 0
        val_num = 0
        _, loss_val_1, accura = cnn_forward_prop(grayscale_img_var, label)
        loss += loss_val_1
        val_num += accura
        loss_array.append(loss / 100)
        accuracy_array.append(val_num)
        dictionary_cont_dat[label] = val_num
    #print(accuracy_array)
    #print(dictionary_cont_dat)
   #print(class_val[max(dictionary_cont_dat, key=dictionary_cont_dat.get)])
    return class_val[max(dictionary_cont_dat, key=dictionary_cont_dat.get)]
actual = []
predicted = []
for directory_name, _, files in os.walk('dataset/dataset_eye/test'):
    for file in files:
        img_loc = os.path.join(directory_name, file)
        classes = img_loc.split("\\")[-2]
        actual.append(classes)
        b_g_im = image.load_img(img_loc,target_size = (48,48),color_mode = "grayscale")
        b_g_im = np.array(b_g_im)
        b_g_im = np.expand_dims(b_g_im,axis = 0)
        b_g_im = b_g_im.reshape(1,48,48)
        prediction = predict(b_g_im)
        predicted.append(prediction)
print("****** report on test data Python numpy Scratch********************************")
print(accuracy_score(actual,predicted))
print(confusion_matrix(actual,predicted))
print(classification_report(actual,predicted))
```
